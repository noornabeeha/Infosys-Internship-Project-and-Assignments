```python
# 2. Build Threat Scoring Logic
#score = 5
# credential	+3
# execution	+2
# privilege	+3
# persistence	+2
# lateral	+2
print('--------------------------------')
print('---------function starts----------')
# for Keyword (in name)	Score Increase, name is the key present in 'techniques' a list of dict
def threat_score(tech):
    score = 5
    if "credential" in tech:
        score+=3
    if "execution" in tech:
        score+=2
    if "privilege" in tech:
        score+=3
    if "persistence" in tech:
        score+=2
    if "lateral" in tech:
        score+=2
    return score

for i in range(800):
    riskname = (threat_score(techniques[i]['name']))
    print(techniques[i]['name'],"--->",riskname)
```

```python
# 'description'

def threat_score_desc(desc):
    score = 5
    if "administrator" in desc:
        score+=2
    if "remote" in desc:
        score+=2
    if "bypass" in desc:
        score+=2
    if "stealth" in desc:
        score+=1
    return score
for i in range(100):
    des = (threat_score_desc(techniques[i]['description']))
    print(techniques[i]['name'],"--->",des)
```

```python
#combining both A and B
def threat_score_name_desc(tech,desc):
    score = 5
    if "credential" in tech:
        score+=3
    if "execution" in tech:
        score+=2
    if "privilege" in tech:
        score+=3
    if "persistence" in tech:
        score+=2
    if "lateral" in tech:
        score+=2
    if "administrator" in desc:
        score+=2
    if "remote" in desc:
        score+=2
    if "bypass" in desc:
        score+=2
    if "stealth" in desc:
        score+=1
    return score
scored_techniques = []
for i in range(len(techniques)):
    sc = threat_score_name_desc(techniques[i]['name'],techniques[i]['description'])
    scored_techniques.append(sc)

#----------------------------------
len(scored_techniques)
```

```python
for i in range(len(techniques)):
    print(techniques[i]['name'],"--->",scored_techniques[i])

#creating a dictionary which can have both values, so when i sort the values for future use, it does not mess up with the indices between both the lists 
# which are 'techniques' and 'scored_techniques'
d={}
for i in range(len(techniques)):
    d[techniques[i]['name']]=scored_techniques[i]

    
```

```python
# 4. Prioritize Threats
# •	Sort techniques from highest score to lowest.
# •	Display the Top 10 highest-risk techniques.

#sorting my dictionary with scores given using previous functions and storing it in a new dict
di = dict(sorted(d.items(), key = lambda x: x[1],reverse = True))

for i, (key, value) in enumerate(di.items()):
    if i >= 10:
        break
    print(key, "--SCORE-->", value)


```

```python
# 5. Critical Threat Filter
# Print only techniques where:
# score >= 8.9
# These represent Critical Threats that analysts should investigate first.
# ________________________________________
for i, (key, value) in enumerate(di.items()):
    if value>8.9:
        print(key, " IS CRITICAL--SCORE-->", value)
        
```
